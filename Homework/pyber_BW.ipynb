```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "/Users/bryanwood/Desktop/MatPlotLib_Homework/Homework/city_data.csv"
ride_data = "/Users/bryanwood/Desktop/MatPlotLib_Homework/Homework/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

# Combine the data into a single dataset
merge_table = pd.merge(city_df, ride_df, on="city", how="outer")
# Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain x and y values for each city type

# Pull data for Urban City type
urban_df = merge_table.loc[(merge_table["type"] == "Urban")]
# Group Urban data by city
urban_group = urban_df.groupby("city")
# Calculate urban drivers per city
urban_driver_per_city = urban_group["ride_id"].count()
# Calculate Urban average fare
urban_average_fare = urban_group["fare"].mean()
# Calculate Urban total rides per city
urban_total_ride_per_city = urban_group["driver_count"].max()

# ---------------------------------------

# Pull data for Suburban City type
suburban_df = merge_table.loc[(merge_table["type"] == "Suburban")]
# Group Suburban data by city
suburban_group = suburban_df.groupby("city")
# Calculate Suburban drivers per city
suburban_driver_per_city = suburban_group["ride_id"].count()
# Calculate Suburban average fare
suburban_average_fare = suburban_group["fare"].mean()
# Calculate Suburban total rides per city
suburban_total_ride_per_city = suburban_group["driver_count"].max()

# ---------------------------------------

# Pull data for Rural City type
rural_df = merge_table.loc[(merge_table["type"] == "Rural")]
# Group Rural data by city
rural_group = rural_df.groupby("city")
# Calculate Rural drivers per city (x-axis)
rural_driver_per_city = rural_group["ride_id"].count()
# Calculate Rural average fare (y-axis)
rural_average_fare = rural_group["fare"].mean()
# Calculate Rural total rides per city (bubble size)
rural_total_ride_per_city = rural_group["driver_count"].max()

# Plot Urban City data points
urban_plot = plt.scatter(urban_driver_per_city, urban_average_fare,s=(urban_total_ride_per_city)*10, label="Urban",marker="o", alpha = .75,facecolors="coral", edgecolors="black")
plt.grid(b=bool)

# Plot Suburban City data points
suburban_plot = plt.scatter(suburban_driver_per_city, suburban_average_fare,s=(suburban_total_ride_per_city)*10, label="Suburban", marker="o", alpha = .75,facecolors="lightskyblue", edgecolors="black")
plt.grid(b=bool)

# Plot Rural City data points
rural_plot = plt.scatter(rural_driver_per_city, rural_average_fare,s=(rural_total_ride_per_city)*10, marker="o", label = "Rural", alpha = .75, facecolors="gold", edgecolors="black")
plt.grid(b=bool)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
legend = plt.legend(title="City Types",loc=1, handles=[urban_plot,suburban_plot,rural_plot])
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

plt.gcf().text(.92, 0.7, "Note: Circle size correlates with driver count per city.", fontsize=12)
# Incorporate a text label regarding circle size

# Save Figure
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban = merge_table.loc[merge_table["type"] == "Urban"]
urban_count = urban["fare"].sum()
suburban= merge_table.loc[merge_table["type"] == "Suburban"]
suburban_count = suburban["fare"].sum()
rural = merge_table.loc[merge_table["type"] == "Rural"]
rural_count = rural["fare"].sum()
total = merge_table["fare"].sum()

urban_per = urban_count / total
suburban_per = suburban_count / total
rural_per = rural_count / total

#sizes
sizes = [urban_per,rural_per,suburban_per]

labels = ["Urban", "Rural", "Suburban"]

# The colors of each section of the pie chart
colors = ["lightcoral", "gold", "lightskyblue"]

explode = (0.1,0.0,0.0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, autopct="%1.1f%%", startangle=280)

plt.show()

# Save Figure
```

## Total Drivers by City Type

```python
# Calculate Type Percents
urban = merge_table.loc[merge_table["type"] == "Urban"]
urban_group = urban.groupby(["city"]).first()
urban_count = urban_group["driver_count"].sum()

suburban = merge_table.loc[merge_table["type"] == "Suburban"]
suburban_group = suburban.groupby(["city"]).first()
suburban_count = suburban_group["driver_count"].sum()

rural = merge_table.loc[merge_table["type"] == "Rural"]
rural_group = rural.groupby(["city"]).first()
rural_count = rural_group["driver_count"].sum()

total = urban_count + suburban_count + rural_count 

urban_per = urban_count / total
suburban_per = suburban_count / total
rural_per = rural_count / total

#sizes
sizes = [urban_per,rural_per,suburban_per]

labels = ["Urban", "Rural", "Suburban"]

# The colors of each section of the pie chart
colors = ["lightcoral", "gold", "lightskyblue"]

explode = (0.1,0.0,0.0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, autopct="%1.1f%%", startangle=240)

plt.show()
```
